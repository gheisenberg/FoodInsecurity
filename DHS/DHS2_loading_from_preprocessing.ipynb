```python
folder_path = "/mnt/datadisk/data/surveys/DHS_final_raw_data/"
work_dir = "/mnt/datadisk/data/Projects/water/pickles/"
input_dir = "/mnt/datadisk/data/Projects/water/inputs/"
min_version = 3
overwrite_pqt = False
parquet = False
use_dask = False
drop_threshold = False
drop_col_counter = False
cols_to_load_simultaneously = 10
dataset_type = 'HR'  #, 'BR', 'CR', 'HW', 'IR', 'KR', 'MR', 'PR', 'AR'] ??['AN', 'FC, 'FP', 'HW', 'PV', 'SC'] < 25 surveys
add_str = 'pd'
columns = []
verbose = False
available_since_v = 7
preprocessing_answers_file = f"{input_dir}summary_df_V3_{dataset_type}__unify_answers_reviewed2b.csv"
# out_f = f"{input_dir}DF_V3_{dataset_type}_after_replacing.pkl"
preprocessing_f = f"{input_dir}Preprocessing_df_V3_{dataset_type}.pkl"
unify_columns_f = f"{input_dir}summary_df_V3_{dataset_type}_unify_cols_reviewed.csv"
cut_perc = 10
```

```python
# Load column names from all .sav files
file_paths = get_dhs_files(folder_path, dataset_type, min_version)
file_paths = sorted(file_paths)
ic(len(file_paths))
```

```python
unify_cols_df = pd.read_csv(unify_columns_f, encoding='ISO-8859-1')
unify_cols_df
```

```python
preprocessing_f_unify_answers = preprocessing_f[:-4] + '_unify_answers.pkl'
if not os.path.exists(preprocessing_f) or overwrite_pqt:
    df_actual_unified_answers, processed_cols = preprocessing_unified_columns(unify_cols_df, file_paths, dataset_type, folder_path, cut_perc)
    df_actual_unified_answers.to_csv(preprocessing_f_unify_answers[:-4] + '.csv', index=False)
    df_actual_unified_answers.to_pickle(preprocessing_f_unify_answers)
    
    # save processed columns to pkl
    with open(preprocessing_f_unify_answers[:-4] + 'processed_cols.pkl', 'wb') as f:
        pickle.dump(processed_cols, f)
else:
    df_actual_unified_answers = pd.read_pickle(preprocessing_f_unify_answers)
    # load processed columns from pkl
    with open(preprocessing_f_unify_answers[:-4] + 'processed_cols.pkl', 'rb') as f:
        processed_cols = pickle.load(f)
df_actual_unified_answers
```

```python
preprocessing_answers_df = pd.read_csv(preprocessing_answers_file, encoding='ISO-8859-1')
preprocessing_answers_df
```

```python
# remove already processed columns
preprocessing_answers_df_loading = preprocessing_answers_df[~preprocessing_answers_df['column_name'].isin(processed_cols)]
preprocessing_answers_df
```

```python
if not os.path.exists(preprocessing_f) or overwrite_pqt:
    preprocessing_answers_df_loading = preprocessing_answers_df_loading[(preprocessing_answers_df_loading["available_data [%]"] >= cut_perc) | 
                                                                (preprocessing_answers_df_loading["available_data V7+ [%]"] >= cut_perc)]
    df = preprocessing_loading_wrapper_wrapper(file_paths, dataset_type, folder_path, cols_to_load_simultaneously, preprocessing_answers_df_loading)
    df.to_csv(preprocessing_f[:-4] + '.csv', index=False)
    df.to_pickle(preprocessing_f)
else:
    df = pd.read_pickle(preprocessing_f)
df
```

```python
# merge with already processed columns
df = df.merge(df_actual_unified_answers, on='DHSID + HHID', how='outer')
df
```
